```python
model = tf.keras.models.load_model('models/Model_006.h5')
model.summary()
```

```python
def create_generator(directory, shuffle=True):
  return tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255.)\
  .flow_from_directory(directory,
                       shuffle=shuffle,
                       target_size=(512,512),
                       batch_size=32,
                       class_mode='binary')
```

```python
train_generator = create_generator(images_path+'/nih')
```

```python
result = model.predict(train_generator)
```
